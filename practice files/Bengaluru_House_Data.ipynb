```python
import pandas as pd 
import numpy as np
```

```python
df = pd.read_csv(r"dataset\Bengaluru_House_Data.csv")
df.head()
```

```python
df.shape
```

```python
unique_values_dict = {}

# Loop through each column and store unique values in the dictionary 
for column in df.columns: 
    unique_values_dict[column] = df[column].unique() 

# Print the dictionary 
for column, unique_values in unique_values_dict.items(): 
    print(f"Unique values in '{column}': {unique_values}\n")
```

```python
df.isnull().sum()
```

```python
df.isnull().sum().sum()
```

 # Filling Null Values

```python
df2 = df.fillna(value = 0)
df2
```

```python
df2.isnull().sum().sum()
```

## fill null value with previous row value

```python
df3 = df.fillna(method = 'pad')
df3
```

```python
df3.isnull().sum().sum()
```

# fill null value with the next row value

```python
df4 = df.fillna(method = 'bfill')
df4
```

# fill with the previous column value

```python
df5 = df.fillna(method = 'pad', axis = 1)
df5
```

# fill with the next column value

```python
df6 = df.fillna(method = 'bfill', axis = 1)
df6
```

# filling different values in different columns

```python
df8 = df.fillna({'society': 'abcd', 'balcony' : 'defg'})
df8
```

# filling with the mean of a column

```python
df9 = df.fillna(value=df['balcony'].mean())
df9
```

# dropna()

```python
df10 = df.dropna()
df10
```

```python
df11 = df.dropna(how='all')
# if all, it will a row where na is all of the values in a row
df11
```

# replace()

```python
df12 = df.replace(to_replace = np.nan, value = 1245)
df12
```
